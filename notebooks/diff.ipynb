```python
import numpy as np
import h5py
import pandas as pd
```

```python
Illustrissnaps = np.arange(0,136,1)
# TNGsnaps = [99, 50, 33, 25, 21]


for snap in Illustrissnaps:
#     print(f"Illustris (snapshot {snap})")
    for size in ["dwarf","massive"]:
        try:
            f = h5py.File(f"../data/groups/Illustris_{snap}.hdf5","r")
            numDark = len(f["dark"][size]["Group Number"])
            numHydro = len(f["hydro"][size]["Group Number"])
            difference = abs(numDark - numHydro)
            print(f"snapshot {snap} {size}:{numDark},{numHydro},{difference}")
        except KeyError:
            print(f"{snap} DNE")
```
